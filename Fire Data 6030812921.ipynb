เลือกข้อมูลที่ต้องการผ่านชื่อของคอลัมน์

```python
dataset = fire_data[['OBJECTID','LATITUDE','LONGITUDE','ACQ_DATE','ACQ_TIME']]
```

สร้าง list เก็บข้อมูลวัน และดูจำนวนข้อูลวันที่ 5 ถึงวันที่ 9

```python
DATE = []
```

```python
Count = 0
for a in dataset['ACQ_DATE']:
    if str(a)[9] == '0':
        break
    if int(str(a)[9]) > 4:
        DATE.append(str(a)[0:10])
        Count+=1
print(DATE)
print(Count)
```

สร้างฟังก์ชันเอาไว้ใช้ตอนสร้างที่เก็บข้อมูล Lat, Long, TIME

```python
def Create_data(name_l, DT):
    name_l = []
    count = 0
    for i in dataset[DT]:
        if count == 589 :
            break
        name_l.append(i)
        count += 1
    return name_l
```

```python
Lat = Create_data('Lat', "LATITUDE")
```

```python
Long = Create_data('Long', "LONGITUDE")
```

สร้าง dict ไว้เก็บค่าวันเวลา z เป็นค่าที่ได้จากการแปลง datetime เป็นตัวเลข แล้วนำไปพล็อตในกราฟ

```python
import datetime as dt
import matplotlib.dates as md
Calender = {}
z = list()
for i in range(len(TIME)): # เริ่มนับที่ 0
    print(i)
    hr=str()
    minute=str()
    
    if len(str(TIME[i])) == 3:
        hr = str(TIME[i])[0]
        minute = str(TIME[i])[1:]
        
    if len(str(TIME[i])) == 4:
        hr =str( TIME[i])[0:2]
        minute = str(TIME[i])[3:5]
                     
    Calender[i] = dt.datetime(int(DATE[i][:4]), int(DATE[i][5:7]), int(DATE[i][8:]), int(hr), int(minute))
    
    z.append(md.date2num(Calender[i]))
    
```

```python
Calender, z
```

พล็อตกราฟโดยให้แกน z แสดงออกเป็นวันเวลา

```python
import matplotlib.pyplot as mplt
import matplotlib.dates as md

fig = mplt.figure(figsize=(15,15))
ax = fig.gca(projection='3d')

ax.scatter(Lat, Long, z, zdir='z', s=40, c = 'red', depthshade = True)
zlabel = md.DateFormatter('%Y-%m-%d %H:%M:%S')
ax.zaxis.set_major_formatter(zlabel)

ax.set_xlabel('Latitude')
ax.set_ylabel('Longitude')
ax.set_zlabel('Date,Time')
```
